```python
from mongo import MongoDBReader

db = MongoDBReader()
```

```python
user_feedbacks = db.get_user_feedbacks()

for feedback in user_feedbacks:
    print(feedback)
```

```python
import matplotlib.pyplot as plt

# Convert cursor to list and extract ratings
ratings = [feedback['rating'] for feedback in db.get_user_feedbacks()]

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(ratings, bins=range(1, 7), align='left', rwidth=0.8)
plt.title('Distribution of User Feedback Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(range(1, 6))
plt.grid(True, alpha=0.3)
plt.show()
```
