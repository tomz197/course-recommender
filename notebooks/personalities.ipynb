```python
# print the names of the people, their studies, and their interests, liked courses, and disliked courses
def print_person_info(person):
    print(f"Name: {person['name']}")
    print(f"Studies: {person['studies']}")
    print(f"Interests: {', '.join(person['interests'])}")
    for liked_course in person['liked_courses']:
        if liked_course in course_names_list:
            print(f"Liked course: {liked_course}")
        else:
            print(f"Liked course: {liked_course} (not found in course names)")
    for disliked_course in person['disliked_courses']:
        if disliked_course in course_names_list:
            print(f"Disliked course: {disliked_course}")
        else:
            print(f"Disliked course: {disliked_course} (not found in course names)")
    print()

for person in people[:1]:
    pass
    print_person_info(person)
```

```python
# Add liked_courses_codes and disliked_courses_codes keys to each person
import pandas as pd

# Get the course names and codes from the courseClient dataframe
course_df = courseClient.df[["NAME_EN", "CODE"]]
course_dict = dict(zip(course_df["NAME_EN"], course_df["CODE"]))

# Function to find course codes based on course names
def find_course_codes(course_names):
    codes = []
    for course_name in course_names:
        if course_name in course_dict:
            codes.append(course_dict[course_name])
        else:
            # If exact match not found, try to find a close match
            for db_course_name in course_dict.keys():
                if course_name.lower() in db_course_name.lower() or db_course_name.lower() in course_name.lower():
                    codes.append(course_dict[db_course_name])
                    break
            else:
                print(f"Course '{course_name}' not found in course names.")
                codes.append(None)  # No match found
    return codes

# Add codes to each person
for person in people:
    person["liked_courses_codes"] = find_course_codes(person["liked_courses"])
    person["disliked_courses_codes"] = find_course_codes(person["disliked_courses"])
```

```python
from google import genai
from google.genai import types
import json

from dotenv import load_dotenv
load_dotenv()

# Initialize the client
client = genai.Client(
    api_key=os.getenv("GEMINI_API_KEY"),
)

def prompt_gemini(input) -> str:
    model = "gemini-2.0-flash"

    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=40,
        max_output_tokens=8192,
        safety_settings=[
            types.SafetySetting(
                category="HARM_CATEGORY_HARASSMENT",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_HATE_SPEECH",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_SEXUALLY_EXPLICIT",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_DANGEROUS_CONTENT",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_CIVIC_INTEGRITY",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            ),
        ],
        response_mime_type="application/json",
        response_schema=genai.types.Schema(
            type = genai.types.Type.OBJECT,
            enum = [],
            required = ["like"],
            properties = {
                "like": genai.types.Schema(
                    type = genai.types.Type.BOOLEAN,
                )
            },
        ),
        system_instruction=[
            types.Part.from_text(text="""# You are a university student. You'll be given this information:
- Your interests
- Your field of study
- List of courses you liked
- List of courses you disliked

You will be given a course name and you will have to decide if you like it or not. You will answer with "like" if you like the course and "dislike" if you don't like it.
"""),
        ],
    )

    contents = [
        types.Content(
            role="user",
            parts=[types.Part.from_text(text=str(input))],
        ),
    ]

    # Collect the response text from streaming
    response_text = ""
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        response_text += chunk.text

    return response_text
```

```python
from main import db

async def test_person_model(person, model="embeddings_mmr"):
    for i in range(3):
        print(f"Test {i+1} for {person['name']}")
        interests = ", ".join(person["interests"])
        studies = person["studies"]
        liked_courses = ", ".join(person["liked_courses"])
        disliked_courses = ", ".join(person["disliked_courses"])

        recommended_course = await recommendations(person["liked_courses_codes"], person["disliked_courses_codes"], [], 1, model, relevance=0.8)
        recommended_course = recommended_course.recommended_courses[0]
        recommended_course_name = recommended_course.NAME_EN
        input = f"""
Interests: {interests}
Studies: {studies}
Liked courses: {liked_courses}
Disliked courses: {disliked_courses}

Course to rate: {recommended_course_name}
    """
        response = prompt_gemini(input)
        response_parsed = json.loads(response)
        liked = response_parsed["like"]
        print(f"Response: {liked} for course {recommended_course_name}")
        
        db.log_bot_recommendation_feedback(
            liked=person["liked_courses_codes"],
            disliked=person["disliked_courses_codes"],
            skipped=[],
            course=recommended_course.CODE,
            action="like" if liked else "dislike",
            user_id=person["name"],
            model=model,
        )
        
        # add the recommended course to liked or disliked courses
        if liked:
            person["liked_courses"].append(recommended_course_name)
            person["liked_courses_codes"].append(recommended_course.CODE)
        else:
            person["disliked_courses"].append(recommended_course_name)
            person["disliked_courses_codes"].append(recommended_course.CODE)

await test_person_model(people[0])
```
