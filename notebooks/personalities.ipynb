```python
from google import genai
from google.genai import types
import json

from dotenv import load_dotenv
load_dotenv()

# Initialize the client
client = genai.Client(
    api_key=os.getenv("GEMINI_API_KEY"),
)

def prompt_gemini(input) -> str:
    model = "gemini-2.0-flash"

    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=40,
        max_output_tokens=8192,
        safety_settings=[
            types.SafetySetting(
                category="HARM_CATEGORY_HARASSMENT",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_HATE_SPEECH",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_SEXUALLY_EXPLICIT",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_DANGEROUS_CONTENT",
                threshold="BLOCK_NONE",  # Block none
            ),
            types.SafetySetting(
                category="HARM_CATEGORY_CIVIC_INTEGRITY",
                threshold="BLOCK_LOW_AND_ABOVE",  # Block most
            ),
        ],
        response_mime_type="application/json",
        response_schema=genai.types.Schema(
            type = genai.types.Type.OBJECT,
            enum = [],
            required = ["like"],
            properties = {
                "like": genai.types.Schema(
                    type = genai.types.Type.BOOLEAN,
                )
            },
        ),
        system_instruction=[
            types.Part.from_text(text="""You are a university student. You'll be given this information:
- Your interests
- Your field of study
- List of courses you liked
- List of courses you disliked

You will be given a course name and description and you will have to decide if you like it or not. You will answer with "like" if you like the course and "dislike" if you don't like it.
You should be conservative about youre like, only like those that match you interests, hobbies and field of study. If you are not sure, answer with "dislike". 

Be VERY PICKY. The course must match your interests very closely. DO NOT like courses that ore only slightly related to your field of study!
"""),
        ],
    )

    contents = [
        types.Content(
            role="user",
            parts=[types.Part.from_text(text=str(input))],
        ),
    ]

    # Collect the response text from streaming
    response_text = ""
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        response_text += chunk.text

    return response_text
```

```python
import concurrent.futures
import threading
import asyncio

# Function to run tests for a batch of people
async def test_batch_of_people(batch_people):
    for person in batch_people:
        for model_name in await models():
            await test_person_model(person, model=model_name)
    return f"Completed batch with {len(batch_people)} people"

# Function to create a new event loop and run the async function
def run_in_thread(batch_people):
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    return loop.run_until_complete(test_batch_of_people(batch_people))

# Divide people into batches of 10
def create_batches(people_list, batch_size=10):
    return [people_list[i:i + batch_size] for i in range(0, len(people_list), batch_size)]

# Run the tests in parallel using ThreadPoolExecutor
def run_parallel_tests(people_list, max_workers=None):
    batches = create_batches(people_list)
    print(f"Created {len(batches)} batches of people")
    
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = [executor.submit(run_in_thread, batch) for batch in batches]
        
        for future in concurrent.futures.as_completed(futures):
            try:
                result = future.result()
                print(result)
            except Exception as e:
                print(f"An error occurred: {e}")

# Run the parallel tests
run_parallel_tests(people)
```
