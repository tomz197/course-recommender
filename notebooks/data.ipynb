```python
import os
import json

all_data = []

# Iterate through all files in the data directory
for root, dirs, files in os.walk('../data/formatted'):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(root, file)
            with open(file_path, 'r', encoding='utf-8') as f:
                faculty_data = json.load(f)
                all_data.extend(faculty_data)
                print(f"Loaded {file}: {len(faculty_data)} records")

print(len(all_data))
```

```python
code_name_faculty_data = []
for course in all_data:
    code_name_faculty_data.append({
        'CODE': course['CODE'],
        'NAME': course['NAME'],
        'FACULTY': course['FACULTY']
    })

for course in code_name_faculty_data:
    print(course)

with open('../data/formatted/all_code_name_faculty.json', 'w', encoding='utf-8') as f:
    json.dump(code_name_faculty_data, f)
```
